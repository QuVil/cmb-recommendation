```python
import numpy as np
import pandas as pd
```

```python
FILE_NAME = "../../data/artist_streams.csv"
SAVE_NAME = "../../data/artist_songs.csv"
```

```python
print("Reading data...")
data = pd.read_csv(FILE_NAME)
print(f"Data shape: {data.shape[0]} lines, {data.shape[1]} columns")
```

```python
data
```

```python
data_songs = pd.read_json('../../data/songs.json', lines=True)

# Keep only songs and artists
data = data[["sng_id", "art_id"]]
data_songs = data_songs[["sng_id", "art_id"]]

# Merge both frames for maximum info
data = data.append(data_songs).drop_duplicates()
print(f"Entries after merge:  {data.shape[0]}")
data
```

```python
data.sort_values("art_id").to_csv(SAVE_NAME, index=False)
```
