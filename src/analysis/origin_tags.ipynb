```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import nbimporter
import mainstream_tests
# from matplotlib import rc
# rc('text', usetex=True)

sns.set_palette(sns.color_palette("colorblind"))
sns.set_style("ticks")
```

```python
FILE_NAME = "../../data/artist_streams.csv"

print("Reading data...")
data = pd.read_csv(FILE_NAME)
data = mainstream_tests.clean_data(data)
```

```python
tags = data["origin"].value_counts(normalize=True) * 100
tags
```

```python
f, ax = plt.subplots(figsize=(7, 7))

sns.barplot(x=tags.index, y=tags.values)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()

plt.savefig("../../data/origin_distrib.pdf")
```
